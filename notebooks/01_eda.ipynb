#### Missing Value Strategy (decided in EDA)

GROUP 1 — NaN means "doesn't exist" → fill with "None" or 0
- PoolQC, MiscFeature, Alley, Fence, FireplaceQu → "None"
- GarageType, GarageFinish, GarageQual, GarageCond → "None"
- GarageYrBlt, GarageCars, GarageArea → 0
- BsmtQual, BsmtCond, BsmtFinType1 → "None"
- BsmtExposure, BsmtFinType2 → "None" (except row 332 and 948)

GROUP 2 — Genuinely missing → fill intelligently
- Row 332: BsmtFinType2 → fill with mode ("Unf")
- Row 948: BsmtExposure → fill with mode ("No")
- LotFrontage → fill with median per neighborhood
- MasVnrArea → fill with 0
- MasVnrType → fill with "None"
- Electrical → fill with mode

## EDA Findings Summary

### Missing Values
- GROUP 1 (NaN = doesn't exist): PoolQC, MiscFeature, Alley, 
  Fence, FireplaceQu, GarageType, GarageFinish, GarageQual,
  GarageCond, GarageYrBlt → fill with "None" or 0
- GROUP 2 (genuinely missing): LotFrontage → median per neighborhood,
  Electrical → mode, MasVnrType → "None", MasVnrArea → 0
- Row 332: BsmtFinType2 → fill with mode ("Unf")
- Row 948: BsmtExposure → fill with mode ("No")

### Target Variable
- SalePrice is right skewed
- Will apply log1p transformation before modeling

### Outliers
- Remove 2 houses: GrLivArea > 4000 AND SalePrice < 200000
- Both in Edwards neighborhood, built 2007-2008 (housing crash)

### Multicollinearity
- GarageCars ↔ GarageArea (0.88) → create combined feature
- TotalBsmtSF ↔ 1stFlrSF (0.82) → create TotalSF feature
- GrLivArea ↔ TotRmsAbvGrd (0.83) → keep GrLivArea, drop TotRmsAbvGrd

```python
from scipy import stats
cat_cols = train.select_dtypes(include=['object']).columns

f_scores={}

for col in cat_cols:
    groups = [train[train[col]==cat]['SalePrice'].values for cat in train[col].unique()]

    f_stat, p_value = stats.f_oneway(*groups)
    f_scores[col] = f_stat

f_scores_df = pd.DataFrame({
    'f_score': f_scores
}).sort_values('f_score', ascending=False)

print(f_scores_df.head(10))
```
